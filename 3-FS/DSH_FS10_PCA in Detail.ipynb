```python
import pandas as pd
import numpy as np
import  seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn import datasets, metrics
from sklearn.model_selection import train_test_split
```

```python
cancer = datasets.load_breast_cancer()
```

```python
print(cancer.DESCR)
```

```python
df = pd.DataFrame(cancer.data, columns=cancer.feature_names)
df.head()
```

```python
df.shape
```

```python
# Before Applying the PCA we need to scale the data i.e standardize the data to same scale.
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

X_scaled = scaler.fit_transform(df)

X_scaled
```

# Apply PCA

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2, random_state=42)

pca.fit(X_scaled)

PCA(n_components=2, random_state=42)
```

```python
X_pca = pca.transform(X_scaled)
```

```python
X_scaled.shape, X_pca.shape

# we can see that 30 features has reduced to 2 features. 
```

```python
plt.figure(figsize=(12,8))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c = cancer.target, cmap = 'viridis')
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.title('Scatter plot for Second principal component and First principal component')
plt.show()

# From the plot we can observe, first principal component has high variance compared to second principal component.
```

```python
pca.explained_variance_ratio_

# we can see PCA1 has more variance. 
```

```python
#  Variance v/s number of PCA

pca = PCA(n_components=20, random_state=42) 
# here 20 means we are trying to see how much variance is captured from PCA1 to PCA20

X_pca = pca.fit_transform(X_scaled)

variance = pca.explained_variance_ratio_

plt.ylabel('Variance')
plt.xlabel('Principal components')
plt.title('Bar graph for variances of the componets ')
plt.bar(x = range(1, len(variance)+1), height=variance, width=0.7)
plt.show()
```
